```python
left = DataFrame({
    'A': [1, 2, 3],
    'B': [6, 7, 8],
    'key': ['k0', 'k1', 'k2']
})
```

```python
left
```

```python
right = DataFrame({
    'A': [11, 22, 33],
    'B': [66, 77, 88],
    'key': ['k0', 'k1', 'k2']
})
```

```python
right
```

```python
merge(left, right, how='inner', on='key') # how is inner by default.
```

```python
left1 = DataFrame({
    'key1':['KO','KO','K1','K2'],
    'key2':['k0','K1','K0','K1'],
    'A':[1,2,3,4],
    'B':[5,6,7,8]
})
```

```python
left1
```

```python
right1 = DataFrame({
    'key1':['KO','K1','K1','K2'],
    'key2':['k0','K0','K0','K0'],
    'C':[11,22,33,44],
    'D':[55,66,77,88]
})
```

```python
right1
```

```python
merge(left1,right1,how='inner',on=['key1','key2'])
```

```python
merge(left1,right1,how='outer',on=['key1','key2'])
```

```python
merge(left1,right1,how='left',on=['key1','key2'])
```

```python
merge(left1,right1,how='right',on=['key1','key2'])
```
